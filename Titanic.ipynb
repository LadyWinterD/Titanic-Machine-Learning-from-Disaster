```python
#impoort data
import numpy as np
import pandas as pd
train = pd.read_csv('C:\\Users\Jenny\\Documents\GitHub\\Titanic-Machine-Learning-from-Disaster\\train.csv')
# test data
test = pd.read_csv('C:\\Users\\Jenny\\Documents\\GitHub\\Titanic-Machine-Learning-from-Disaster\\test.csv')
print('Data trainning:',train.shape,'test data:',test.shape)
```

```python
rowNum_train=train.shape[0]
rowNum_test=test.shape[0]
print('tranning data has ',rowNum_train,'rows')
print('test data has',rowNum_test,'rows')
```

```python
#combine the data
full = train.append( test , ignore_index = True )
print('after combined data:',full.shape)
```

```python
#review data
full.head()
```

```python
full.describe()
#full describetion
```

```python
# check each type of data and amount
full.info()
```

```python
#There are 1306 rows,
# Integer data:
# 1, the total rows of age is 1046, it missed 1309 - 1046 = 263, 263/1309=20%
# 2, The total of Fare is 1308, it miies one ,  
# String data:
#1, Embarked 1037 ---> miised 2 
#The total of Cabain is 295, 1309-292 = 1014 , it miised 1014/1309 = 77.5%

# For integer data, it can be replaced by average data
# For classify data, it can be replaced by poplar data
# For missing value, it can be replaced by Model predicts
print('before process')
full.info()
#age
full['Age'] = full['Age'].fillna(full['Age'].mean())
#fare
full['Fare'] = full['Fare'].fillna(full['Fare'].mean())
print('after process')
full.info()
```

```python
#check the data after filling value
full.head()
```

```python
#string data
#Embarked
# departure S = Southampton
# stop station c = Cherbourg
# Stop station Q= Queenstown
full['Embarked'].head()
```

```python
# from the result Southampton is a high frequency value, so missing value could be the hight frequency value
full['Embarked'].value_counts()
```

```python
full['Embarked'] = full['Embarked'].fillna('S')
#for Cabin , the missing value will be filled by U(Unknow)
full['Cabin'] = full['Cabin'].fillna('U')
full.head()
```

```python
full.info()
```

```python
#Data classify-Sex:Male,Female,Embarked,Pclass,Name,Cabin,Ticked
# integer data : PassengerID, Age, Fare, SibSp, Parch
full['Sex'].head()
```

```python
#Set up the sex
def sex_value(Sex):
    if Sex=='male':
        return 1
    else:
        return 0
full['Sex'] = full['Sex'].apply(lambda x:sex_value(x))
full.head()
```

```python
#use dummies to covert it to one-hot code, it will make 
embarkedDf = pd.DataFrame()
embarkedDf = pd.get_dummies( full['Embarked'],prefix='Embarked')
embarkedDf.head()
```

```python
#add one-hot code to the dummy variables and drop Embarked
full = pd.concat([full,embarkedDf],axis=1)
full.drop('Embarked',axis=1,inplace=True)
full.head()
```

```python
#PClass
#1= First Class
#2 = Second Class
#3 = Third Class
PclassDF = pd.DataFrame()  
PclassDF= pd.get_dummies(full['Pclass'],prefix='Pclass')
PclassDF.head()
```

```python
# Add the Class to the data modeling
full = pd.concat([full,PclassDF],axis=1)
#Dell Class
full.drop('Pclass',axis=1,inplace=True)
full.head()
```

```python
# Name
#Extract the title from names  
def getTitle(name):
    str1=name.split(',')[1]    #Heikkinen,Miss.Laina
    str2=str1.split('.')[0]      
    str3=str2.strip() #remove the space
    return str3
#save the value
titleDf = pd.DataFrame()
#calculate for each series
titleDf['Title'] = full['Name'].map(getTitle)
titleDf.head()
```

```python
title_mapDict={
                    "Col":        "Officer",
                    "Major":      "Officer",
                    "Jonkheer":   "Royalty",
                    "Don":        "Royalty",
                    "Sir" :       "Royalty",
                    "Dr":         "Officer",
                    "Rev":        "Officer",
                    "the Countess":"Royalty",
                    "Dona":       "Royalty",
                    "Mme":        "Mrs",
                    "Mlle":       "Miss",
                    "Ms":         "Mrs",
                    "Mr" :        "Mr",
                    "Mrs" :       "Mrs",
                    "Miss" :      "Miss",
                    "Master" :    "Master",
                    "Lady" :      "Royalty"
    
}
titleDf['Title'] = titleDf['Title'].map(title_mapDict)
titleDf=pd.get_dummies(titleDf['Title'])
titleDf.head()
```

```python
full =  pd.concat([full,titleDf],axis=1)
#delete name
full.drop('Name',axis=1,inplace=True)
full.head()
```

```python
#Catin
CabinDf= pd.DataFrame()
full['Cabin'] = full['Cabin'].map(lambda c :c[0])
cabinDf = pd.get_dummies( full['Cabin'] , prefix = 'Cabin' )
cabinDf.head()
```

```python
full = pd.concat([full,cabinDf],axis=1)
full.drop('Cabin',axis=1,inplace=True)
full.head()
#add cabin to the data modeling
```

```python
# building the family and classify
#Family_single: Family has one member
#Fmily_Small: Family has 2 members but less than 4
#Fmily_Large:Family has more than 5 members
familyDf= pd.DataFrame()
familyDf[ 'FamilySize' ] = full[ 'Parch' ] + full[ 'SibSp' ] + 1
familyDf[ 'Family_Single' ] = familyDf[ 'FamilySize' ].map( lambda s : 1 if s == 1 else 0 )
familyDf[ 'Family_Small' ]  = familyDf[ 'FamilySize' ].map( lambda s : 1 if 2 <= s <= 4 else 0 )
familyDf[ 'Family_Large' ]  = familyDf[ 'FamilySize' ].map( lambda s : 1 if 5 <= s else 0 )
familyDf.head()
```

```python
full = pd.concat([full,familyDf],axis=1)
full.head()
```

```python
#Calculate Classify 
corrDf = full.corr()
corrDf 
corrDf['Survived'].sort_values(ascending=False)
```

```python
# feature choose
full_X = pd.concat([titleDf,
                   PclassDF,
                   familyDf,
                   full['Fare'],
                   cabinDf,
                   embarkedDf,
                   full['Sex']
                   ],axis=1 )
full_X.head()
```

```python
source_X = full_X.loc[0:890] #-1 because it starts from [0]
source_y = full.loc[0:890,'Survived'] 
pred_X = full_X.loc[0:890,:]
print('There are ',source_X.shape,'rows in the original source')
print('There are ',pred_X.shape,'rows in the predicted source')
```

```python
from sklearn.model_selection import train_test_split 

# build data modeling
train_X,test_X,train_y,test_y = train_test_split(source_X,
                                                source_y,
                                                train_size=.8)
print('Original source feature:',source_X.shape,
     'Training data feature:',train_X.shape,
     'Test data feature:',test_X.shape)
print('Original source tag:',source_y.shape,
     'Training data tag:',train_y.shape,
     'Test data tag:',test_y.shape)
```

```python
from sklearn.linear_model import LogisticRegression
#Import the linear_model
model = LogisticRegression()
#Train the model
# import warnings filter
from warnings import simplefilter
# ignore all future warnings
simplefilter(action='ignore', category=FutureWarning)
```

```python
model.fit(train_X,train_y)
```

```python
model.score(test_X,test_y)
```

```python
pred_Y = model.predict(pred_X)
```

```python
pred_Y=pred_Y.astype(int)
```
